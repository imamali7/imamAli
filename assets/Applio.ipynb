```python
import os
import requests
from tqdm import tqdm

# 1. الإعدادات
API_KEY = "AIzaSyBTZZzK67MLmrxWTxWTY4Byx5Z4pwVVUHg"
FOLDER_ID = "1Tfz1s9WYBAI8lv2rbA4_lzVg6Txv_piJ"
DATASET_PATH = "/kaggle/working/Applio/datasets/dataset"

os.makedirs(DATASET_PATH, exist_ok=True)

def get_files_in_folder(folder_id, api_key):
    """جلب قائمة الملفات داخل المجلد باستخدام Google Drive API"""
    # استعلام لجلب الأسماء، المعرفات، ونوع الملف
    url = f"https://www.googleapis.com/drive/v3/files?q='{folder_id}'+in+parents&fields=files(id,name,mimeType)&key={api_key}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json().get('files', [])
    else:
        print(f"❌ خطأ في الاتصال: {response.status_code}")
        print(response.text)
        return []

def download_file(file_id, file_name, destination, api_key):
    """تحميل ملف واحد مع شريط تقدم"""
    url = f"https://www.googleapis.com/drive/v3/files/{file_id}?alt=media&key={api_key}"
    dest_path = os.path.join(destination, file_name)
    
    # استخدام stream=True للتعامل مع الملفات الكبيرة دون استهلاك الرام
    response = requests.get(url, stream=True)
    total_size = int(response.headers.get('content-length', 0))
    
    with open(dest_path, 'wb') as f, tqdm(
        desc=f"📥 {file_name}",
        total=total_size,
        unit='iB',
        unit_scale=True,
        unit_divisor=1024,
    ) as bar:
        for data in response.iter_content(chunk_size=32768): # حجم قطعة البيانات 32 كيلوبايت
            size = f.write(data)
            bar.update(size)

# 2. التنفيذ
print("📡 جاري فحص محتويات المجلد عبر API...")
files = get_files_in_folder(FOLDER_ID, API_KEY)

if not files:
    print("⚠️ لم يتم العثور على ملفات. تأكد من أن المجلد يحتوي على ملفات وأن صلاحية المشاركة مفعلة (Anyone with the link).")
else:
    print(f"📂 تم العثور على {len(files)} ملف. بدء التحميل الآن...")
    for file_info in files:
        # تحميل الملفات فقط وتجاهل المجلدات الفرعية مؤقتاً
        if file_info['mimeType'] != 'application/vnd.google-apps.folder':
            try:
                download_file(file_info['id'], file_info['name'], DATASET_PATH, API_KEY)
            except Exception as e:
                print(f"❌ فشل تحميل {file_info['name']}: {e}")
        else:
            print(f"⏩ تخطي المجلد الفرعي: {file_info['name']}")

print("\n✨ اكتملت المهمة! يمكنك الآن التحقق من مجلد dataset.")
```

```python
import os

def get_exact_file_paths():
    # المجلد الأساسي للبحث
    base_dir = "/kaggle/working/"
    
    # الامتدادات المطلوبة
    audio_exts = ('.wav', '.mp3', '.flac', '.m4a')
    
    print(f"🚀 فحص المسارات الدقيقة للملفات المطلوبة...\n")
    print(f"{'الصنف':<20} | {'الحجم':<10} | {'المسار الكامل'}")
    print("-" * 100)

    found_data = []

    for root, dirs, files in os.walk(base_dir):
        for file in files:
            file_path = os.path.abspath(os.path.join(root, file))
            _, ext = os.path.splitext(file)
            ext = ext.lower()
            
            category = None
            
            # 1. ملفات التدريب المسبق G و D
            if file.startswith(('G_', 'D_')) and ext == ".pth":
                category = "G/D Pretrain"
            
            # 2. ملفات الأوزان النهائية Weights
            elif "/weights/" in root.lower() and ext == ".pth":
                category = "Model Weights"
            
            # 3. ملفات البصمة Index
            elif ext == ".index":
                category = "Index File"
            
            # 4. الملفات الصوتية في Dataset
            elif any(d in root.lower() for d in ["dataset", "datasets"]):
                if ext in audio_exts:
                    category = "Audio Dataset"
            
            if category:
                try:
                    size_mb = os.path.getsize(file_path) / (1024 * 1024)
                    print(f"{category:<20} | {size_mb:>7.2f} MB | {file_path}")
                    found_data.append({"category": category, "path": file_path, "size": size_mb})
                except OSError:
                    continue

    # تحليل المسار الجذري لبرنامج Applio
    print("\n" + "="*50)
    print("📌 توجيهات المسارات لبرنامج Applio:")
    
    applio_path = "/kaggle/working/Applio"
    if os.path.exists(os.path.join(applio_path, "Applio")):
        final_root = os.path.join(applio_path, "Applio")
        print(f"⚠️ تنبيه: لديك مجلد مزدوج. المسار الرئيسي لملفاتك هو:\n{final_root}")
    else:
        final_root = applio_path
        print(f"✅ المسار الرئيسي لبرنامجك هو:\n{final_root}")
    
    print("\n💡 نصيحة: انسخ المسار الكامل للملف الذي تريده من القائمة أعلاه")
    print("وضعه في خانة المسار المناسبة في واجهة Applio.")
    print("="*50)

    return found_data

# تشغيل الفحص وعرض النتائج
all_paths = get_exact_file_paths()
```
